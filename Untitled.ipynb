```python
import pandas as pd
import json
import re
import nltk
from nltk.corpus import stopwords
from gensim import corpora, models
from textblob import TextBlob

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# 1. Load JSON Data 
def load_youtube_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    # only "title" and "time" columns use and 'Watched ' prefix remove
    df = df[df['title'].str.contains('Watched ', na=False)]
    df['clean_title'] = df['title'].str.replace('Watched ', '', regex=False)
    return df

# 2. Text Cleaning Function
def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE) # Remove URLs 
    text = re.sub(r'\W', ' ', text) # Remove Special characters 
    text = text.lower() # Lowercase
    words = text.split()
    words = [w for w in words if not w in stop_words and len(w) > 2] # Remove Stopwords & short words 
    return words

# 3. Sentiment Analysis
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Main Execution
file_path = 'watch-history.json' 
df = load_youtube_data(file_path)

# Cleaning 
df['tokenized_title'] = df['clean_title'].apply(clean_text)

# 4. Topic Modeling (LDA)
# Dictionary and Corpus 
dictionary = corpora.Dictionary(df['tokenized_title'])
corpus = [dictionary.doc2bow(text) for text in df['tokenized_title']]

# LDA Model 
lda_model = models.LdaModel(corpus, num_topics=4, id2word=dictionary, passes=15)

# Logic for categorization of Topics
def assign_topic(tokens):
    bow = dictionary.doc2bow(tokens)
    topic_probs = lda_model.get_document_topics(bow)
    return sorted(topic_probs, key=lambda x: x[1], reverse=True)[0][0]

df['topic_id'] = df['tokenized_title'].apply(assign_topic)
df['sentiment'] = df['clean_title'].apply(get_sentiment)

# Save final data
df[['time', 'clean_title', 'topic_id', 'sentiment']].to_csv('youtube_enriched_data.csv', index=False)

print("Enriched data saved to youtube_enriched_data.csv")
```
